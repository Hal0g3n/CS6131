### Section A: Overview & Business Rules

### Overview

<div class="alert alert-block alert-warning">
Give a short background of your proposed domain, and why a RDBMS system is appropriate.
</div>


Chess has been gaining popularity since the start of the pandemic. With the game of chess becoming more mainstream, someone needs to provide the service of playing chess online. A chess website is one of the ways of providing such services, with several purposes such as:
- **Online play**: Chess websites provide an easy and convenient way for people to play chess against other players from around the world.
- **Learning and Improvement**: Chess websites offer resources such as puzzles, and tactics trainers to help users improve their skills.
- **Community**: Chess websites provide a community for chess players to interact and connect with others who share their interest in the game.
- **Competition**: Chess websites offer organized tournaments and events for players to compete against each other and test their skills.
- **Accessibility**: Chess websites make the game of chess accessible to people who may not have easy access to physical chess sets or opponents.

When making a chess website with the above purposes, there is a lot of data to store, each relatad to one another. Tournament information, Player information, Player history, Played Games, Puzzles, Tactics. These are all data that the website must handle. With a Relational DataBase Management System (RDBMS), the data can be managed easily and efficiently, allowing for a smooth experience when playing chess online. 

### Business Rules

<div class="alert alert-block alert-warning">
Complete your writeup of the final business rules here. The business rules writeup should be AS CLEAR AS possible. Someone reading your writeup should be able to clearly model a similar ER you have proposed in Section B.
</div>


A **Player** must be either an **AI** or **Human**, and has a unique ***username*** and ***history*** of their chess rating. \
The ***history*** is a list of elo ***rating*** and the ***datetime***. ***history*** is only updated on rating change. \
Each **Human** player has a ***email***, ***birthday***, ***country*** and ***password*** for login purposes, while the **AI** has the link to their ***api***. \
**AI**'s chess rating will remain stagnant, while a **Human**'s chess rating can change, with the default rating being 1000.

2 **Player**s will play a **Game**. The **Game** has a ***time control***, ***pgn id*** to locate the pgn file storing the game and ***is public*** or not. \
A public **Game** is one where the game can be viewed by anyone. \
***time control*** will consist of ***initial time*** and ***increment*** per move.

A **Player** will join a **Room** to play a **Game**. \
A **Room** has a unique 4 character ***id***, and will have up to 2 players in it, but at least 1. \
If a **Room** has 0 players, it will be deleted automatically.

A **Team** has a unique ***name***, ***creation date*** and **Players** can *belong to* a **Team**. \
On joining, the ***join date*** of the **Player**s are recorded.
**Team**s are moderated by some **Human** players in the team, and their ***start date*** is recorded. \
A **Team** must have at least 1 **Human** moderating and the creator is automatically a moderator of the team. If there are no moderators, the most active member will become a moderator. \
If no members are found (ie there are no players in the **Team**), then the **Team** is deleted.


A **Player** can create an **Application**, containing a short ***message*** and an ***id*** unique for each player. \
The **Application** is to join an existing **Team**. \
Only a **Team**'s moderator can approve the application.


**Tournament**s are hosted by a **Team** and only **Player**s in the **Team** can *participate* in the **Tournament**. \
Each **Tournament** hosted will have a unique ***id***, ***start datetime***, ***end datetime***, ***time control***, and whether it ***is rated***. \
**Tournament**s can be **Private** or **Public**, and has **Game**s *under* it. \
When **Player**s *participate*, their ***ranking***, raw ***score*** and ***performace*** will be saved.

**Puzzle**s can be *attempted* by **Player**, and the **Player** ***is correct*** or not, taking some ***time***. \
***time*** is a factor in the rating change of the player, and will be added to the history.
**Puzzle**s has a unique ***id***, ***rating***, ***movelist*** of puzzle solution, ***fen*** of position and ***solve count***.

Inheritance Mapping:
- **AI** and **Human** inherits **Player**: Subclass relations only. This is because there are very little AI entities, and storing in the same table will lead to many null values, wasting space.
  - As a result of this, ***player1_name*** and ***player2_name*** could also refer to the ***username*** in the AI table.
  - This is represented by the second arrow from ***player1_name*** and ***player2_name*** pointing to ***username*** in the AI table.

Relationship Mapping:
- one-to-many: Following the norm, we will store the foreign key in the entity with cardinality many. This reduces number of tables, and the number of joins when performing a query. \
  The relationships are in this category are:
  - `Creates`
  - `Approves`
  - `Apply_to`
  - `Hosts`
  - `Part_of`

<br/>

- many-to-many: Following the norm, we will store the relationship in a new table. This is the only way to adequately represent the relationship. \
  The relationship is in this category are:
  - `Attempts`
  - `Participates`

<br/>

- `Plays` (one-to-one-to-many): This is a ternary relationship, with 2 **Player**s and 1 **Game** entity. Just like the one-to-many relationship, we store the both **Player**s' ***username*** (foreign key) in the **Game** entity. \
  This is also space efficient since **Game** undergoes total participation in this relationship.


- `Moderates` and `Belong_to`: There is a special condition for this relationship, it is that the **Team** the **Player** has the relationship with is the same for both `Moderates` and `Belong_to`. \
  Thus, we can combine the foreign key into ***team_name***. To determine which relationship is formed (or both), we look at the attributes ***mod_start_date*** for `Moderates` and ***join_date*** for `Belong_to`. \
  If they are non-null, then the relationship exists. \
  This reduces the number of entries we need to encapsulate both relationships.

Multi-valued Mapping:
- History: Each entry is made up of ***rating*** and ***datetime*** (could not be represented in ER model). This is stored in a separate table history.
